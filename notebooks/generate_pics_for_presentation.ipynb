```python
import matplotlib.pyplot as plt
import json
import numpy as np
%config Completer.use_jedi = False
```

```python
with open("results.json") as jfile:
    data = json.load(jfile)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data, columns=["N","python","vec_python","segyio","i2f","numba","cupy"])
```

```python
cols = df.columns[1:]
for col in cols:
#     plt.plot(df["N"],np.log10(df[col]*1000000), label=col)
    plt.plot(df["N"],df[col], label=col)
plt.legend()
plt.xlabel("Number of traces converted")
plt.ylabel("total time (s)")
plt.title("Comparing read speeds of various methods")
plt.savefig("Time read 1000 traces.png",format='png')
```

```python
cols = df.columns[1:]
for col in cols:
    plt.plot(df["N"],np.log10(df[col]*1000000), label=col)
#     plt.plot(df["N"],df[col], label=col)
plt.legend(bbox_to_anchor=(1,0), loc="lower right", ncol=3)
plt.xlabel("Number of traces converted")
plt.ylabel("Log10 total time (us)")
plt.title("Comparing read speeds of various methods (log time)")
plt.savefig("Time read 1000 traces logtime.png",format='png')
```

```python
cols = df.columns[1:]
for col in cols[3:]:
    plt.plot(df["N"],np.log10(df[col]*1000000), label=col)
#     plt.plot(df["N"],df[col], label=col)
plt.legend(bbox_to_anchor=(1,0), loc="lower right", ncol=3)
plt.xlabel("Number of traces converted")
plt.ylabel("Log10 total time (us)")
plt.title("Comparing read speeds of three fastest methods (log time)")
plt.savefig("Time read 1000 traces 3 mthods logtime.png",format='png')
```
